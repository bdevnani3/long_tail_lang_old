```python
import json
```

```python
LT_data = "/nethome/bdevnani3/flash1/long_tail_lang/data/ImageNet_LT/ImageNet_LT_train.txt"
```

```python
count = {}
with open(LT_data, "r") as f:
    for line in f:
        _, label = line.split(" ")
        if int(label) not in count:
            count[int(label)] = 0
        count[int(label)] +=1
        
classification = {
    "low": [],
    "med": [],
    "high": [],
}

for k, v in count.items():
    if v <= 20:
        classification["low"].append(k)
    elif v <= 100:
        classification["med"].append(k)
    else:
        classification["high"].append(k)
        
print(classification)

with open("/nethome/bdevnani3/flash1/long_tail_lang/data/ImageNet_LT/freq_classification.json", "w") as f:
    json.dump(classification, f)
```
